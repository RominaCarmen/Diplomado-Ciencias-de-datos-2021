# Elements of a Plot

## Initial Configuration

```python
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use("bmh")
plt.rcParams["figure.figsize"] = 16, 5
```

## Load Data

```python
df = sns.load_dataset("anscombe")
```

```python
df.groupby("dataset").describe(percentiles=[]).T
```

## Different Figsizes

```python
np.random.seed(42)
random_walk = np.random.normal(size=1000).cumsum()

figsizes = [
    (4, 4),
    (6, 4),
    (8, 4),
    (10, 4),
    (12, 4),
    (14, 4),
    (16, 4),
    (18, 4),
    (20, 4),
    (22, 4),
    (24, 4),
]

for figsize in figsizes:    
    plt.figure(figsize=figsize)
    plt.plot(random_walk)
    plt.title(f"{figsize=}")
    plt.show()
```

## Encoding Information

### Facet

```python
fix, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 10), sharex=True, sharey=True)
axes = np.ndarray.flatten(axes)

columns = ("I", "II", "III", "IV")

for column, ax in zip(columns, axes):
    data = df[df.dataset == column]
    sns.scatterplot(x="x", y="y", data=data, s=120, ax=ax)
    ax.set_title(column)

plt.tight_layout()
```

### Color

```python
sns.scatterplot(x="x", y="y", hue="dataset", data=df, s=120)
plt.tight_layout()
```

### Shape

```python
sns.scatterplot(x="x", y="y", style="dataset", data=df, s=120)
plt.tight_layout()
```

### Size

```python
sns.scatterplot(x="x", y="y", size="dataset", size_order=["IV", "III", "II", "I"], data=df, sizes=[50, 100, 150, 200])
plt.tight_layout()
```

### Shape and Color

```python
sns.scatterplot(x="x", y="y", hue="dataset", style="dataset", data=df, s=120)
plt.tight_layout()
```
