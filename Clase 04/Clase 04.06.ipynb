## Exercise 1

Plot the following dataset using points and color each variable differently.

```python
size = 1000

np.random.seed(20210929)

dataset = pd.DataFrame({
    't': np.arange(size),
    'x': np.random.normal(1, 0.5, size=size).cumsum(),
    'y': np.random.normal(1, 0.5, size=size).cumsum(),
})
```

```python
dataset
```

```python
alt.Chart(dataset).mark_point().encode(
    x='x',
    y='y',
    color="t"
).interactive()
```

## Exercise 2

Add a moving avarage to the previous graph with a windows size of 20, without including future observations

```python
alt.Chart(dataset).mark_point().encode(
    x='y',
    y='x',
    color="t"
).transform_window(
    moving_average='mean(x)',
    frame=[-20, 0],
)
```

## Exercise 4

Use some imputation rule to fill the gaps in the dataset and plot it with the original data in the background with low opacity

```python
size = 1000

np.random.seed(20210929)

dataset = pd.DataFrame({
    't': np.arange(size),
    'x': np.random.normal(1, 0.5, size=size).cumsum(),
    'y': np.random.normal(1, 0.5, size=size).cumsum(),
})

x_index, y_index = np.random.random(size=(2, len(dataset))) < 0.1

dataset.loc[x_index, "x"] = np.nan
dataset.loc[y_index, "y"] = np.nan

dataset.head()
```
