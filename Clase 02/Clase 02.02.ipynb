```python
import numpy as np

from matplotlib import pyplot as plt
from matplotlib import animation, rc
import colorsys

rc('animation', html='html5')
plt.rcParams["figure.figsize"] = 16, 4
plt.style.use("bmh")
```

## Function to animate

```python
def plot_frame(frame, step_proportion, ax):
    ax.cla()
    step = frame / step_proportion
    x = np.arange(step, 2 * np.pi + step, 0.05)
    y = np.sin(x)
    ax.plot(x, y)
    ax.set_yticks([])
    ax.set_xticks([])
    return []
```

## Basic Animations

```python
figure, ax = plt.subplots();
figure.set_tight_layout(True)
anim = animation.FuncAnimation(figure, plot_frame, fargs=(5, ax,))
anim
```

## Saving Animations

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=5)

anim.save('output/native_basic.mp4', writer=writer)
```

## Animation Smoothness

### 5 FPS

```python
figure, ax = plt.subplots();
figure.set_tight_layout(True)

anim = animation.FuncAnimation(figure, plot_frame, frames=31, fargs=(5, ax,))

plt.close()
```

```python
anim
```

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=5)

anim.save('output/native_5fps.mp4', writer=writer)
```

### 10 FPS

```python
figure, ax = plt.subplots();
figure.set_tight_layout(True)

delay = 1 / 10
duration = np.pi * 2
frames = round(duration / delay)
anim = animation.FuncAnimation(figure, plot_frame, frames=frames, interval=delay*1000, fargs=(10, ax,))

plt.close()
```

```python
anim
```

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=10)

anim.save('output/native_10fps.mp4', writer=writer)
```
