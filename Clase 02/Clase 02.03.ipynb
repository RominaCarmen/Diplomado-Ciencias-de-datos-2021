# Animations With Celluloid

```python
from pathlib import Path
Path("output").mkdir(exist_ok=True)
```

```python
import numpy as np

from matplotlib import pyplot as plt
from matplotlib import animation, rc

from celluloid import Camera  # !pip install celluloid

rc('animation', html='html5')
plt.rcParams["figure.figsize"] = 16, 4
plt.style.use("bmh")
```

## Basic Animations

```python
figure, ax = plt.subplots();
figure.set_tight_layout(True)

camera = Camera(figure)

ax.set_yticks([])
ax.set_xticks([])

for i in range(30):
    step = i / 5
    x = np.arange(0, 2 * np.pi, 0.05)
    y = np.sin(x + step)
    ax.plot(x, y, color="steelblue")
    camera.snap()

animation = camera.animate()
```

```python
animation
```

## Saving Animations

```python
animation.save('output/celluloid_basic.mp4')
```

## Animation Smoothness

### 5 FPS

```python
figure, ax = plt.subplots();
figure.set_tight_layout(True)

camera = Camera(figure)

ax.set_yticks([])
ax.set_xticks([])

fps = 5
delay = 1 / fps
duration = np.pi * 2
frames = round(duration / delay)

for i in range(frames):
    step = i / fps
    x = np.arange(0, 2 * np.pi, 0.05)
    y = np.sin(x + step)
    ax.plot(x, y, color="steelblue")
    camera.snap()

animation = camera.animate(interval=delay*1000)
plt.close()
```

```python
animation
```

```python
animation.save('output/celluloid_5fps.mp4')
```
